1. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
import random
rand_list=list(np.random.randint(low = 1,high=100,size=50))

with open('lista_random.txt', mode='w') as f:
    for i in range(len(rand_list)):
        f.write(str(rand_list[i]))
        f.write('\n')
```

2. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
matrix_rand = np.random.randint(0,100, size=(5,5))
print(matrix_rand)

with open('matrix_random.txt', mode='w') as f:
    for i in range(matrix_rand.shape[0]):
        f.write(' '.join(map(str, matrix_rand[i])))
        f.write('\n')
        
```

3. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
with open("matrix_random.txt", "r") as f:
    data = f.read().replace(" ", ",")
    
    
with open("matrix_random.csv", "w") as g:
    g.write(data)
```
